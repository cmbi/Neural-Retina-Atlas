```python
import pandas as pd
```

```python
bed = pd.read_csv('../data/transcriptomics/isoquant_aln.sorted.transcript_models.bed12', sep = '\t', header= None)
bed.columns = ['chrom', 'chromStart', 'chromEnd', 'name', 'score', 'strand', 'tickStart', 'thickEnd', 'itemRgb', 'blockCount', 'blockSizes', 'blockStarts']
bed.head()
```

```python
# Define a custom function to map names to colors
def map_name_to_color(name):
    if 'nnic' in name:
        return '175,62,77'
    elif 'ENST' in name:
        return '78,128,152'
    elif 'nic' in name:
        return '231,111,81'
    else:
        return None

bed['itemRgb'] = bed['name'].apply(map_name_to_color)

bed.to_csv('../data/transcriptomics/isoquant_aln.sorted.transcript_models_colored.bed12', sep = '\t', header=False, index = False)

bed.head()

```
